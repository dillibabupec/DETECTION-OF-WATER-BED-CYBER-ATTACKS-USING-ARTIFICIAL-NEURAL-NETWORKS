```python
# Import the neccessary packages
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy
```

```python
# Set the display max columns datasets
pd.set_option('display.max_columns',None)
```

```python
# Read the dataset
data = pd.read_csv('A.csv')
```

```python
# Check the top values
data.head()
```

```python
del data['DATETIME']
del data['F_PU9']
del data['S_PU9']
del data['F_PU5']
del data['S_PU5']
del data['F_PU3']
del data['S_PU3']
del data['S_PU1']
```

```python
# Remove the null values
df1 = data.dropna()
```

```python
df1.head()
```

```python
df = df1.rename({'Stress Level':'Stress_level'},axis=1)
```

```python
# Check the total number of null values.
df.isnull().sum()
```

```python
df.head(20)
```

```python
# Check the data relationships
# df.head(20)
```

```python
# Check the unique values in specific columns
df['ATT_FLAG'].unique()
```

```python
# Split the datasets into traing process X is (independent features, independent features)
x1 = df.drop(labels='ATT_FLAG',axis=1)
y1 = df.loc[:,'ATT_FLAG']
```

```python
x1.head(20)
```

```python
x1.info()
```

```python
# Balance the datasets
import imblearn
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

ros =RandomOverSampler(random_state=1)
x,y=ros.fit_resample(x1,y1)
print("OUR DATASET COUNT         : ", Counter(y1))
print("OVER SAMPLING DATA COUNT  : ", Counter(y))
```

```python
x.info()
```

```python
# Split the datasets into training and testing
from sklearn.model_selection import train_test_split

x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.50) 
```

```python
# Check the unique values
y.unique()
```

```python
# Import the model build algorithm
from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier()
clf.fit(x_train, y_train)
predictRFC =clf.predict(x_test)
```

```python
# Check the accuracy score of this model
from sklearn.metrics import accuracy_score

accuracy_score = accuracy_score(y_test, predictRFC)* 100
print("Accuracy Score\n\n", accuracy_score)
```

```python
# Check the number True positive value, True negative value of this model
from sklearn.metrics import confusion_matrix

conf = confusion_matrix(y_test, predictRFC)
print("Confusion matrix\n\n", conf)
```

```python
# Check the classification report of this model
from sklearn.metrics import classification_report

report = classification_report(y_test, predictRFC)
print("Classification Report\n\n",report)
```

```python
# Save the model for deployment purpose.
import joblib
joblib.dump(clf, 'RFE.pkl')
```
